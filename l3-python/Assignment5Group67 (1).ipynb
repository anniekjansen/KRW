### Semi-Formal Relationships

This is the list of all semiformal relationships for the new model

####RELATION 1: Generating beliefs from observing

AT EACH TIME STEP, 

IF ambient agent X observes Y

THEN ambient agent X will believe Y at the next time step


####RELATION 2: Observing the level of alcoholism of A

AT EACH TIME STEP, 

IF A has a level of alcoholism L

THEN ambient agent X will observe that A has a level of alcoholism L at the next time step

####RELATION 3: Observing the chance of cardiovascular disease of A

AT EACH TIME STEP, 

IF A has a level of cardiovascular disease L

THEN ambient agent X will observe that A has a level of cardiovascular disease L at the next time step


####RELATION 4: Belief of A's feeling of loneliness

AT EACH TIME STEP, 

IF ambient agent X believes that A has a level of cardiovascular disease "high" AND a level of alcoholism "medium"

OR

IF ambient agent X believes that A has a level of cardiovascular disease "medium" AND a level of alcoholism "high"

THEN, at the next time step, ambient agent X will believe that A feels loneliness 


####RELATION 5: Assessing the that A is in a desiderable state in regards to A's believed loneliness (according to the agent X)

AT EACH TIME STEP,

IF ambient agent x believes that A feels loneliness

AND

IF ambient agent x believes that feeling loneliness is a desiderable state for A

THEN, at the next time step, ambient agent x will assess that A is in a desidarable state in regards A's loneliness (belived by X)
